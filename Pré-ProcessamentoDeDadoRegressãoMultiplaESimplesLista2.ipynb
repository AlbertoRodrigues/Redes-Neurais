# Observando os tipos das variáveis dadas pelo R.

```r
#Percebi que os tipos não estão todos certos, como mencionado anteriormente o dado faltante foi primeiramente dado por "?"
#, assim o R automaticamente deduziu que é uma variável categórica mas isso não é sempre verdade, e nos próximos blocos 
#de código consertarei esse problema.
a=c()
for(i in 1:ncol(dados2))
{
  x=cbind(class(dados2[,i]),colnames(dados2)[i])
  a=rbind(a,x)
}
a
```

# Verificando as variáveis que o R considerou como variável numérica, ou seja, as variáveis que não apresentam dados faltantes, consequentemente não apresentaram "?" no inicio ocasionando interpretação do R como variável numérica.

```r
v=c()
for(i in 1:ncol(dados2))
{
  if(is.numeric(dados2[,i])||is.integer(dados2[,i]))
  {
    v=append(v,i)
  }
}
v
table(is.na(dados[,v]))
```

# Os Indices das Variáveis numéricas reais do banco de dados, logo fazendo a diferença entre o conjunto com todos os indices das variáveis numéricas e os indices das variáveis que o R considera como numérica teremos os indices das variáveis que faltam fazer alterações no seu tipo.

```r
#Todos os indices das variáveis numéricas.
x=c(1,2,10,11,12,13,14,17,19,20,21,22,23,24,25,26);x
#Essas variáveis não estão sendo utilizadas como variáveis numéricas pois essas variáveis tinham "?" como dado faltante
# antes de transformar em NA, e o R automaticamente entende essa variável como categórica mesmo ela não sendo, então 
#estamos mudando o tipo dela para numérico.
x=setdiff(x,v);x
##Observando o resumo estatístico dessas variáveis, verifiquei que são realmente variáveis numéricas e o R 
#entende como categórica.
summary(dados2[,x])
#Mudando o tipo dessas variáveis.
for(i in x)
{
  dados2[,i]=as.numeric(as.character(dados2[,i]))
}
```

# Agora fazendo essa alteração vemos que o resumo das medidas estatísticas estão adequados.

```r
summary(dados2[,x])
```

# Verificando todos os indices das variáveis numéricas e um resumo de medidas estatísticas.

```r
v=union(v,x);v
summary(dados2[,v])
```

# Irei fazer ajustes em três variáveis categóricas para os estimadores dos métodos dos mínimos quadrados derem todos finitos.

```r
#Ajuste na variável num-of-cylinders
x=c()
for(i in dados2$`num-of-cylinders`)
{
  if((i=="eight")||(i=="five")||(i=="four")||(i=="three"))
  {
    x=append(x,"eight or five or four or three")
  }
  else
  {
    x=append(x,"six or twelve or two")
  }
}
x
dados2$`num-of-cylinders`=x
dados2$`num-of-cylinders`=factor(dados2$`num-of-cylinders`)
```

```r
#Ajuste na variável fuel-system
x=c()
for(i in dados2$`fuel-system`)
{
  if((i=="1bbl")||(i=="2bbl")||(i=="4bbl")||(i=="idi"))
  {
    x=append(x,"1bll or 2bll or 4bbl or idi")
  }
  else
  {
    x=append(x,"mfi or mpfi or spdi or spfi")
  }
}
x
dados2$`fuel-system`=x
dados2$`fuel-system`=factor(dados2$`fuel-system`)
```

```r
#Ajuste na variável engine-type
x=c()
for(i in dados2$`engine-type`)
{
  if((i=="dohc")||(i=="dohcv")||(i=="l")||(i=="ohc"))
  {
    x=append(x,"dohc or dohcv or l or ohc")
  }
  else
  {
    x=append(x,"ohcf or ohcv or rotor")
  }
}
x
dados2$`engine-type`=x
dados2$`engine-type`=factor(dados2$`engine-type`)
```

# Tratamento dos valores faltantes.
### As variáveis horsepower, peak-rpm, Bore e stroke foram imputados através do algoritmo knn, em que observamos as 10 observações mais próximas através de uma mistura de distâncias, se a variável for numérica a distância será a euclidiana, se a variável for categórica e com valores iguais a distância atribuída é 0, se forem diferentes a distância dada é 1, então pegamos a mediana dos 10 vizinhos mais próximos dessa variável se o valor da variável faltante de cada observação for numérica .
### A Variável num-of-doors por ser categórica pegamos dos 10 vizinhos mais próximos a moda e esse valor é imputado no valor faltante.

```r
table(is.na(dados2))
```

```r
require(DMwR)
#Função de imputação
dados2=knnImputation(dados2,k=10,scale=T,meth="median")
```

```r
table(is.na(dados2))
```

# Modelo de regressão com todas as variáveis possíveis para verificarmos a proporção explicada pela Regressão e depois fazemos uma seleção de variáveis para comparação.

```r
#O modelo explica muito bem com todas as 24 variáveis, cerca de 95,67% a variabilidade da variável resposta é explicada 
#pela regressão, mas será que se reduzirmos uma quantidade desnecessária de variáveis para a previsão o modelo continua 
#explicando bem? Veremos com a seleção de variáveis.
colnames(dados2)

(ajuste=lm(price~symboling+make+`fuel-type`+aspiration+`num-of-doors`+
             `body-style`+`drive-wheels`+`engine-location`+`wheel-base`+length+` width`+height+
             `curb-weight`+`engine-type`+`num-of-cylinders`+`engine-size`+`fuel-system`+bore+
              stroke+`compression-ratio`+horsepower+`peak-rpm`+`city-mpg`+`highway-mpg`,data=dados2))

x=summary(ajuste)
#Proporção da variabilidade explicada pela regressão.
x$r.squared
```

# Seleção de variáveis com o método de regressão  Backward baseada em retirar variável por variável até que todas as variáveis explicativas sejam estatísticamente significativas baseadas pelo critério do valor-p.

```r
step(ajuste,direction="backward")
```

```r
#Com 16 variáveis a variabilidade é explicada pela regressão de 95.53% com AIC=3064.47, havendo um decréscimo de 95,67% - 95.53% = 0.14%
# o que pode ser considerado desprezível e ainda explicando bem o modelo, então será feita essa seleção de variáveis.
(ajuste1=lm(price~`fuel-type`+make+aspiration+`engine-location`+
             `body-style`+`wheel-base`+length+` width`+height+
             `curb-weight`+`engine-size`+`fuel-system`+bore+
             `compression-ratio`+`peak-rpm`+`highway-mpg`,data=dados2))
x=summary(ajuste1);x
x$r.squared
```
